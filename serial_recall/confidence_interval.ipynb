```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fast_recall_results = pd.read_csv('../Data/serial_recall_1_fast05.csv')
slow_recall_results = pd.read_csv('../Data/serial_recall_1_slow4.csv')
```

```python
from scipy.stats import norm

# Function to calculate confidence intervals
def calculate_confidence_interval(data, confidence=0.95):
    z = norm.ppf(1 - (1 - confidence) / 2)  # Z-score for the given confidence level
    data['ci_lower'] = data['proportion'] - z * ((data['proportion'] * (1 - data['proportion'])) / data['total']) ** 0.5
    data['ci_upper'] = data['proportion'] + z * ((data['proportion'] * (1 - data['proportion'])) / data['total']) ** 0.5
    return data

# Apply the function to both datasets
fast_recall_results = calculate_confidence_interval(fast_recall_results)
slow_recall_results = calculate_confidence_interval(slow_recall_results)

# Display the updated DataFrame
fast_recall_results[['proportion', 'ci_lower', 'ci_upper']].head()
```
